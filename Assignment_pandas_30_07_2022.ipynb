# 13.total contribution hour for each and every agents weekly basis

```python
df_agentlogin=pd.read_csv(r"C:\Users\vijay\Downloads\AgentLogingReport (1).csv")
```

```python
df_agentlogin['week_start_new']=pd.to_datetime(df_agentlogin['Date'])-pd.offsets.Week(weekday=4)
```

```python
df_agentlogin['week_end_new']=df_agentlogin['week_start_new']+pd.offsets.Week(weekday=3)
```

```python
df_agentlogin
```

```python
df_agentlogin['Duration_Int']=pd.to_datetime(df_agentlogin['Duration']).apply(lambda x:x.value)
```

```python
df_weeklyConDuration=df_agentlogin.groupby(['Agent',pd.Grouper(key='week_start_new',freq='W')])['Duration_Int'].agg(weeklyContDur='sum')
```

```python
df_weeklyConDuration['weeklyContDur']=pd.to_datetime(df_weeklyConDuration['weeklyContDur']).dt.time
```

```python
df_weeklyConDuration['weeklyContDur'].fillna(0,inplace=True)
```

```python
df_weeklyConDuration
```

```python
df_weeklyConDuration.to_excel(r"C:\Users\vijay\Downloads\weeklyConDuration2.xlsx")
```
